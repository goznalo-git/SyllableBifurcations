```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import scienceplots

plt.style.use(["science","no-latex"])
```

```python
with open("Indices/subsampled_VdP.pkl", "rb") as f:
    indices_VdP = pickle.load(f)

with open("Indices/subsampled_Hopf.pkl", "rb") as f:
    indices_Hopf = pickle.load(f)
    
with open("Indices/subsampled_SNILC.pkl", "rb") as f:
    indices_SNILC = pickle.load(f)
```

```python
xs = {}
ys = {}
zs = {}

for bif, index_dict in zip(["VdP","Hopf","SNILC"],
                           [indices_VdP,indices_Hopf,indices_SNILC]):
    xs[bif] = []
    ys[bif] = []
    zs[bif] = []
    for n, indices in index_dict.items():
        xs[bif].append(indices["center"])
        ys[bif].append(indices["iIRR"])
        zs[bif].append(indices["lam1"]/indices["lam2"])
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1,projection='3d')

##################
### 3D Scatter ###
##################

ax1.scatter(xs["VdP"], ys["VdP"], zs["VdP"], alpha=0.5,
           s=range(20, 20+10*len(xs["VdP"]), 10), c="tab:blue", label="VdP")
ax1.scatter(xs["SNILC"], ys["SNILC"], zs["SNILC"], alpha=0.5,
          s=range(20, 20+10*len(xs["SNILC"]), 10), c="tab:green", label="SNILC")
ax1.scatter(xs["Hopf"], ys["Hopf"], zs["Hopf"], alpha=0.8,
           s=range(20, 20+10*len(xs["Hopf"]), 10), c="tab:red", label="Hopf")

ax1.set_xlabel("center")
ax1.set_ylabel("iIRR")
#ax1.set_zlabel("$\lambda_1/\lambda_2$")
ax1.set_title("$\lambda_1/\lambda_2$")

ax1.ticklabel_format(axis='both', style='sci', scilimits=(0,4))


###########################
### lam1/lam2 vs Center ###
###########################

ax2 = fig.add_subplot(2,2,2)

ax2.scatter(xs["VdP"], zs["VdP"], alpha=0.5,
           s=range(20, 20+10*len(xs["VdP"]), 10), c="tab:blue", label="VdP")
ax2.scatter(xs["SNILC"], zs["SNILC"], alpha=0.5,
          s=range(20, 20+10*len(xs["SNILC"]), 10), c="tab:green", label="SNILC")
ax2.scatter(xs["Hopf"], zs["Hopf"], alpha=0.8,
           s=range(20, 20+10*len(xs["Hopf"]), 10), c="tab:red", label="Hopf")

ax2.set_xlabel("center")
ax2.set_ylabel("$\lambda_1/\lambda_2$")

ax2.ticklabel_format(axis='both', style='sci', scilimits=(0,4))

#########################
### lam1/lam2 vs iIRR ###
#########################

ax3 = fig.add_subplot(2,2,3)

ax3.scatter(ys["VdP"], zs["VdP"], alpha=0.5,
           s=range(20, 20+10*len(xs["VdP"]), 10), c="tab:blue", label="VdP")
ax3.scatter(ys["SNILC"], zs["SNILC"], alpha=0.5,
          s=range(20, 20+10*len(xs["SNILC"]), 10), c="tab:green", label="SNILC")
ax3.scatter(ys["Hopf"], zs["Hopf"], alpha=0.8,
           s=range(20, 20+10*len(xs["Hopf"]), 10), c="tab:red", label="Hopf")

ax3.set_xlabel("iIRR")
ax3.set_ylabel("$\lambda_1/\lambda_2$")

ax3.ticklabel_format(axis='both', style='sci', scilimits=(0,4))

######################
### iIRR vs Center ###
######################

ax4 = fig.add_subplot(2,2,4)
 
ax4.scatter(xs["VdP"], ys["VdP"], alpha=0.5,
           s=range(20, 20+10*len(xs["VdP"]), 10), c="tab:blue", label="VdP")
ax4.scatter(xs["SNILC"], ys["SNILC"], alpha=0.5,
          s=range(20, 20+10*len(xs["SNILC"]), 10), c="tab:green", label="SNILC")
ax4.scatter(xs["Hopf"], ys["Hopf"], alpha=0.8,
           s=range(20, 20+10*len(xs["Hopf"]), 10), c="tab:red", label="Hopf")

ax4.set_xlabel("center")
ax4.set_ylabel("iIRR")

ax4.ticklabel_format(axis='both', style='sci', scilimits=(0,4))

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()


#ax1.view_init(45,45,0)

plt.tight_layout()

plt.savefig("Indices/3d-scatter_synthetic.png")
```
